```python
# Dependencies
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
import numpy as np
import random 
from config import api_key
```

## Generate Cities List and Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
i = 1
num = 500
units = "metric"

cities = [] 
lat = [] 
temperature = []
humidity = []
cloudy = []
wind_speed = []

while i <= num:
#     lon_response = Rand(-180.000, 180.000, 1)
#     lat_response = Rand(-90.000, 90.000, 1)
    lat_response = np.random.uniform(low=-90.000, high=90.000, size=1500)
    lon_response = np.random.uniform(low=-180.000, high=180.000, size=1500)
    
    url_part_1 = "http://api.openweathermap.org/data/2.5/weather?"
    url_part_2 = "lat="
    url_part_3= "&lon="
    url_part_4 = "&appid=9a67f55d9749983852a3a6c6b094885a&units={units}"

    query_url_open_weather = (f'{url_part_1}{url_part_2}{lat_response[0]}{url_part_3}{lon_response[0]}{url_part_4}')
    response = requests.get(query_url_open_weather).json()
    try:
        if(response['name'] != ''):

            cities.append(response['name'])
            lat.append(response['coord']['lat'])
            temperature.append(response['main']['temp'])
            humidity.append(response['main']['humidity'])
            cloudy.append(response['clouds']['all'])
            wind_speed.append(response['wind']['speed'])
            print(f'{i}) The temperature in the city called {cities[-1]} is {temperature[-1]}°F presenting a latitude {lat[-1]}°')

            i += 1
    except:
        print(f'error en {i}')
        
weather_dict = {"City": cities, "Latitude": lat, "Temperature": temperature, "Humidity": humidity, "Cloudy": cloudy, "Wind Speed": wind_speed}
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```

## Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.






## Latitude vs. Temperature Plot

```python
# Build a scatter plot for each data type
plt.scatter(weather_data["Latitude"], weather_data["Temperature"], marker="o")

# Incorporate the other graph properties
plt.title("Temperature (°F) vs. Latitude")
plt.ylabel("Temperature (°F)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure and generate output file
plt.savefig("TemperatureInWorldCities_Temperature_vs_Latitude.png")
weather_data.to_csv("TemperatureInWorldCities.csv", index=False, header=True)

# Show plot
plt.show()
```
